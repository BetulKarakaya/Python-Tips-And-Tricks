```python
import pandas as pd
import numpy as np
from IPython.display import display # Import this to use display() in scripts

class DataVisualizer:
    """
    A class to enhance DataFrame presentation using 
    Pandas' built-in styling engine.
    """
    def __init__(self, data):
        self.df = pd.DataFrame(data)

    def __str__(self):
        return "DataFrame ready for styling."

    def apply_business_styles(self):
        """
        Applies a professional heatmap and highlights 
        maximum/minimum values for quick auditing.
        """
        # TRICK: .style returns a Styler object that can be customized
        styled_df = (self.df.style
            .background_gradient(cmap='Greens', subset=['Profit']) # Heatmap for Profit
            .highlight_max(color='lightgreen', subset=['Revenue']) # Highlight best revenue
            .highlight_min(color='#ff9999', subset=['Costs'])      # Highlight lowest costs
            .format("{:.2f}", subset=['Profit', 'Revenue'])       # Round to 2 decimals
        )
        return styled_df

def main():
    # Sample Dataset: Financial performance of products
    financial_data = {
        'Product': ['Software', 'Hardware', 'Service', 'Cloud', 'Support'],
        'Revenue': [50000, 85000, 30000, 95000, 20000],
        'Costs': [10000, 60000, 15000, 40000, 5000],
        'Profit': [40000, 25000, 15000, 55000, 15000]
    }

    visualizer = DataVisualizer(data=financial_data)
    styled_view = visualizer.apply_business_styles()

    # TRICK: Check if we are in a terminal or a notebook
    try:
        # This will work in Jupyter/Colab
        display(styled_view)
    except NameError:
        # This is the fallback for standard terminal/console
        print("--- Terminal Output (Styles not supported in console) ---")
        print(visualizer.df)

if __name__ == "__main__":
    main()
```
